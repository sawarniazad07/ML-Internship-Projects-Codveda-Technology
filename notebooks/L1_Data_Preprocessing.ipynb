```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Load dataset
df = pd.read_csv("/content/1) iris.csv")
df.head()
```

```python
# Handle missing values
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Encode categorical columns
for col in df.select_dtypes(include='object').columns:
    df[col] = LabelEncoder().fit_transform(df[col])
```

```python
df.head()
```

```python
# Split features & target
X = df.drop("species", axis=1)
y = df["species"]
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Normalize
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print("Preprocessing Done")
```

```python
# Convert arrays back to DataFrames
X_train_df = pd.DataFrame(X_train, columns=X.columns)
X_test_df = pd.DataFrame(X_test, columns=X.columns)

# Add target column
train_df = X_train_df.copy()
train_df["target"] = y_train.values
train_df["split"] = "train"

test_df = X_test_df.copy()
test_df["target"] = y_test.values
test_df["split"] = "test"

# Combine
final_df = pd.concat([train_df, test_df], ignore_index=True)

# Save
final_df.to_csv("/content/iris_dataset_with_split.csv", index=False)
```

```python
final_df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load dataset
df = pd.read_csv("/content/2) Stock Prices Data Set.csv")

# Handle missing values
df.fillna(df.mean(numeric_only=True), inplace=True)

# Encode categorical columns
for col in df.select_dtypes(include='object').columns:
    df[col] = LabelEncoder().fit_transform(df[col])

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Load dataset
df = pd.read_csv("/content/3) Sentiment dataset.csv")
df.head(2)
```

```python
df.drop(columns=["Unnamed: 0.1", "Unnamed: 0"], inplace=True)
df.head()
```

```python
df.info()
```

```python
df.dropna(subset=["Text", "Sentiment"], inplace=True)
df
```

```python
TEXT_COL = "Text"
TARGET_COL = "Sentiment"

le = LabelEncoder()
df[TARGET_COL] = le.fit_transform(df[TARGET_COL])

# ---------------------------
# 4. Train-test split
# ---------------------------
X_train, X_test, y_train, y_test = train_test_split(
    df[TEXT_COL],
    df[TARGET_COL],
    test_size=0.2,
    random_state=42
)

# ---------------------------
# 5. Text Vectorization (TF-IDF)
# ---------------------------
vectorizer = TfidfVectorizer(
    max_features=5000,
    stop_words='english'
)

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

```

```python
# ---------------------------
# 6. Convert to DataFrame (optional for saving)
# ---------------------------
X_train_df = pd.DataFrame(X_train_vec.toarray())
X_test_df = pd.DataFrame(X_test_vec.toarray())

# ---------------------------
# 7. Save everything cleanly
# ---------------------------
# Add labels + split flag
train_df = X_train_df.copy()
train_df["target"] = y_train.values
train_df["split"] = "train"

test_df = X_test_df.copy()
test_df["target"] = y_test.values
test_df["split"] = "test"

final_df = pd.concat([train_df, test_df], ignore_index=True)

# Save dataset
final_df.to_csv("/content/final_sentiment_dataset.csv", index=False)



print("Preprocessing Complete. Dataset Saved.")
```
